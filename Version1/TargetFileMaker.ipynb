```python
import pandas as pd
import numpy as np
import itertools
```

MutliFingerFlexion Adaptation:

Having participatns produce fingerflexions of one digit, two digits, three, up to 5 digits. A random cursor purturbation with zero mean and a certain variance will be added (n_trials = 30). Chords are chosen randomly. They should flex at -2 N.

```python
fingers = [1, 2, 3, 4, 5]

one_finger_chords = [[f] for f in fingers]
two_finger_chords = list(itertools.combinations(fingers, 2))
three_finger_chords = list(itertools.combinations(fingers, 3))
four_finger_chords = list(itertools.combinations(fingers, 4))
five_finger_chords = list(itertools.combinations(fingers, 5))

# Target Chord
target_chord = [-2, -2, -2, -2, -2] # target force for all fingers

iti = 1000 # msecs for inter-trial interval
planTime = 2000 # msecs for precue time
feedbackTime = 2000 # msecs for feedback time


total_sub_num = 2
num_trials_per_block = 30
```

```python
sample_chord = four_finger_chords[0]
sample_chord[2]
```

```python
def chord_forces(chord):
    return [target_chord[i-1] if i in list(chord) else 0 for i in range(1,6)]
```

```python
### target file headers: [subNum, planTime, feedbackTime, iti, session, 
# targetForce1, targetForce2, targetForce3, targetForce4, targetForce5]

np.random.seed(42) # for reproducibility
for sub in range(total_sub_num):
    subNum = sub + 1
    bn = 1
    all_chords = one_finger_chords + two_finger_chords + three_finger_chords + four_finger_chords + five_finger_chords
    # shuffle chords
    np.random.shuffle(all_chords)
    for chord in all_chords:
        chosen_chord = chord_forces(chord)
        block_data = pd.DataFrame([[subNum, chosen_chord[0], chosen_chord[1], chosen_chord[2], chosen_chord[3], chosen_chord[4], 
                                    planTime, feedbackTime, iti]], 
                                    columns = ['subNum', 'targetForce1', 'targetForce2', 'targetForce3', 'targetForce4', 'targetForce5',
                                               'planTime', 'feedbackTime', 'iti'])
        block_data = pd.concat([block_data]*num_trials_per_block, ignore_index=True)
        block_data.to_csv(f'sn{subNum}_{bn}.tgt', index=False, sep='\t')
        bn += 1
    
    
```
