```python
import pandas as pd
import numpy as np
import itertools
```

Single Finger Adaptation

```python
fingers = ['1', '2', '3', '4', '5']

one_finger_chords = [int(f) for f in fingers]

iti = 1000 # msecs for inter-trial interval
planTime = 2000 # msecs for precue time
feedbackTime = 2000 # msecs for feedback time


total_sub_num = 2
num_sessions = 1

num_trials_per_block = 40
```

```python
### target file headers: [subNum, planTime, feedbackTime, iti, session, 
### targetForce1, targetForce2, targetForce3, targetForce4, targetForce5]

for sub in range(total_sub_num):
    subNum = sub + 1
    session = 1
    bn = 1
    for chord in one_finger_chords:
        target_forces = -3
        ### Baseline blocks
        is_target_visible = 0 
        block_data = pd.DataFrame([[subNum, chord, is_target_visible, planTime, feedbackTime, iti, session, target_forces]], columns=['subNum', 'chord', 'targetVisible', 'planTime', 'feedbackTime', 'iti', 'session', 'targetForce'])

        block_data = pd.concat([block_data]*num_trials_per_block, ignore_index=True)
        block_data.to_csv(f'sn{subNum}_bn{bn}.tgt', index = False, sep = '\t')
        bn += 1


        ### Gain 1 blocks 
        is_target_visible = 1
        block_data = pd.DataFrame([[subNum, chord, is_target_visible, planTime, feedbackTime, iti, session, target_forces]], columns=['subNum', 'chord', 'targetVisible', 'planTime', 'feedbackTime', 'iti', 'session', 'targetForce'])
        block_data = pd.concat([block_data]*num_trials_per_block, ignore_index=True)
        block_data.to_csv(f'sn{subNum}_bn{bn}.tgt', index = False, sep = '\t')
        bn += 1

        ### Gain 2 blocks
        is_target_visible = 1
        block_data = pd.DataFrame([[subNum, chord, is_target_visible, planTime, feedbackTime, iti, session, target_forces]], columns=['subNum', 'chord', 'targetVisible', 'planTime', 'feedbackTime', 'iti', 'session', 'targetForce'])
        block_data = pd.concat([block_data]*num_trials_per_block, ignore_index=True)
        block_data.to_csv(f'sn{subNum}_bn{bn}.tgt', index = False, sep = '\t')
        bn += 1


        ### Washout blocks
        is_target_visible = 0
        block_data = pd.DataFrame([[subNum, chord, is_target_visible, planTime, feedbackTime, iti, session, target_forces]], columns=['subNum', 'chord', 'targetVisible', 'planTime', 'feedbackTime', 'iti', 'session', 'targetForce'])
        block_data = pd.concat([block_data]*num_trials_per_block, ignore_index=True)
        block_data.to_csv(f'sn{subNum}_bn{bn}.tgt', index = False, sep = '\t')
        bn += 1


            
```
